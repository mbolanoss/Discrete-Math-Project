#Introducción</br>
Actualmente, la gran mayoría de personas en el mundo hacen uso de distintos softwares como aplicaciones de escritorio, aplicaciones móviles, sitios web, etc, que facilitan diferentes aspectos de su vida cotidiana, estos muy frecuentemente optan por darle la opción al usuario de crearse una cuenta para guardar su información, progreso o cambios que este realice usandolo, y esto conlleva a que cada persona tenga creadas muchas cuentas en múltiples lugares de la internet, cada una asociada a un determinado software que use. Poseer un gran número de cuentas implica, idealmente, tener que pensar en una contraseña distinta para cada cuenta, y esto puede causar que los usuarios olviden sus contraseñas o realicen malas prácticas cómo usar una misma contraseña para todas sus cuentas, o tener anotadas en físico o digitalmente todas sus contraseñas, exponiendolas a que terceros tengan acceso a ellas de una manera sencilla, o que al tener tantas acabe inevitablemente por olvidar alguna y tenga que solicitar su cambio.</br>

Ahora, haciendo uso del algoritmo de cifrado asimétrico RSA (Rivest, Shamir y Adleman), podría implementarse una solución de almacenamiento cloud para aquellas personas con este inconveniente de baja seguridad en sus contraseñas para múltiples sitios web, permitiendo generar una gran base de datos donde las contraseñas de cada usuario se puedan descifrar usando una clave que únicamente la persona propietaria de la cuenta conozca, y que le aporte mayor variabilidad en la complejidad de las contraseñas de otros sitios web, esto generaría una confianza mayor por parte del usuario ya que sabe que nadie más que él va a poder acceder a todo el contenido que se agregue en dicha base de datos, pues ni el mismo administrador de la base de datos donde se alojan las contraseñas encriptadas las podría desencriptar sin ayuda del usuario.</br>

Para lograr desarrollar un gestor de contraseñas que haga uso del algoritmo RSA, se debe primero explicar el funcionamiento y contexto de este algoritmo.El algoritmo RSA fue desarrollado por Ron Rivest, Adi Shamir y Leonard Adleman (científicos del MIT) en 1977 y patentado en 1983. Este es un algoritmo de cifrado asimétrico, es decir, usa un par de llaves o claves, una pública y una privada para cifrar y descifrar el mensaje que se quiere enviar, por lo que cuando el emisor antes de enviar el mensaje, se cifra usando una clave pública del futuro receptor del mensaje, y éste lo descifra usando su clave privada, por lo que la implementación de este algoritmo evita tener que enviar también la clave para descifrar el mensaje. Este algoritmo es uno de los más seguros que existen, ya que se basa en la complejidad computacional de resolver el problema de factorización de números enteros de gran tamaño.</br>

La idea sobre la que se construye el algoritmo, generalmente se realiza planteando la siguiente situación: Hay 2 personas, Bob y Alice, y Bob le quiere enviar un mensaje secreto a Alice, de manera que solo ella lo pueda leer. Para esto, Alice envía a Bob una caja
con un candado abierto, cuya llave sólo la tiene Alice. Luego, Bob recibe la caja, escribe el mensaje y lo guarda dentro de la caja, cerrando el candado. Una vez hecho esto, Bob envía la caja a Alice y ella la abre con su llave. En este caso, la caja y el candado son la clave pública de Alice, y la llave del candado es su clave privada.

Actualmente existen varios gestores de contraseñas bastante robustos como LastPass, Dashlane, KeePass, 1Password, entre otros. Estos software no solo guardan las contraseñas si no que algunos ofrecen funcionalidades extra como la generación de contraseñas seguras o el registro de información de tarjetas de crédito o cuentas bancarias para agilizar las compras por internet. Y en cuanto a algoritmos de encriptación de contraseñas, se han desarrollado varios algoritmos como son los siguientes:
* MD5: este es un algoritmo de reducción criptográfica de 128 bits mayormente usado en el desarrollo web. No se recomienda su uso ya que desde el 2004 se han detectado vulnerabilidades colisiones de hash.
* SHA1: el algoritmo Secure Hash Algorithm es un sistema de funciones hash criptográficas, desarrolladas por la Agencia de Seguridad Nacional de los Estados Unidos. Este algoritmo recibe una entrada cuyo tamaño máximo no puede superar los 264 bits y produce una salida encriptada de 160 bits. Se basa principalmente en el diseño de algoritmos como MD4 o MD5. 
*  SHA2: es el sucesor del SHA1, y este tiene 4 variantes, las cuales dependen del tamaño de la salida que se requiera.
*  AES: el algoritmo Advanced Encryption Standard es un esquema de cifrado por bloques desarrollado por Estados Unidos. Este algoritmo es el más completo y complejo, y cuenta con una *llave privada*, por lo que el algoritmo sirve tanto como para encriptar como para desencriptar.

**Resultados y teoremas usados:**</br>
En el proceso de generación de las claves públic y privada, específicamente en el paso 3, se mencionan 2 resultados previos los cuales son los siguientes:

*   $\phi (p) = p-1$ sí $p$ es primo.
*   Sí $m$ y $n$ son coprimos, entonces $\phi (mn) = \phi (m)\phi (n)$.

Para demostrar el primer resultado, basta con considerar la definición de un número primo, la cual establece que si un número $n$ es primo, sus únicos divisores son el 1 y el mismo, por lo que cualquier número menor a él, es coprimo a él, ya que cualquier número menor no tendrá ningún divisor en común con $n$, por lo que existen $n-1$ números coprimos a $n$.</br>

Y para demostrar el segundo resultado, se parte de la definición de la función Phi de Euler. Así, $\phi (nm) = |\{ a \in \mathbb Z^+; a< mn \ \wedge mcd(nm,a)=1 \} |$. Ahora, $mcd(nm,a) = 1 \Leftrightarrow mcd(n,a) = 1 \ \wedge mcd(m,a)=1$. Ahora se procede a listar todos los numeros de 1 hasta $nm$ dispuestos en filas y columnas, tal que: \\

Ahora se hallarán la cantidad de números en esa matriz son coprimos a $n$ y $m$. Como se puede ver, todo número en la matriz es de la forma $qm + r$, y se sabe que $mcd(qm+r) = mcd(r,m)$, por ejemplo, si se toma la 3ra columna, los números en ella son de la forma $qm+3$ y se cumple que $mcd(qm+3) = mcd(3,m)$. Ahora se plantea la siguiente pregunta: ¿Cuántas columnas hay con primos relativos con $m$?, en efecto, hay $\phi(m)$ columnas. Luego, es natural preguntarse, en cada columna, ¿cuántos números son coprimos a $n$?. Para responder a esta pregunta, se tomará la *r-ésima* columna, y dicha columna es de la forma:</br>
$r$</br>
$m+r$</br>
$2m+r$</br> 
$\vdots$</br>
$(n-1)m+r$</br>
Y se tiene que la columna *r* está formada por enteros coprimos con $m$, además, si $mcd(r,n) = 1$ entonces $mcd(qm+r,n)=1$, esto si $mcd(qm,n) = 1$. Y como $mcd(m,n)=1$ y $0\leq q< n$, si $mcd(q,n)=1 \Rightarrow mcd(qm+r,n)=1$, por lo que habrán tantos $q$ coprimos  con $n$, como valor tenga $\phi (n)$.</br>
Finalmente, en cada columna hay $\phi (n)$ coprimos con $n$, y como hay $\phi (m)$ columnas con números coprimos a $m$, se cumple que en toda la matriz hay $\phi (n) \phi (m)$ coprimos a $nm$. Así: $\phi (nm) = \phi (n) \phi (m)$.

**A continuación se describirán los pasos del RSA:**</br>
Primero se deben generar las claves pública y privada, para esto se realiza el siguiente procedimiento:</br>
1) Se eligen dos números primos $p$ y $q$.</br>
2) Se calcula $n = p\cdot q$.</br>
3) Se calcula $z = \phi(n) = (p-1)\cdot(q-1)$. A esto se le conoce como la función *Phi de Euler*. Esta función se puede evaluar de esta manera ya que se cumple que $\phi (p) = p-1$ sí $p$ es primo, y sí $m$ y $n$ son coprimos, entonces $\phi (mn) = \phi (m)\phi (n)$.</br>
4) Se elige un número primo $k$, tal que $k$ sea co-primo y menor a $z$.</br>
5) La clave pública va a ser el conjunto de los números $(n,k)$.</br>
6) Ahora se calcula la clave privada. Para ello, se elige un número $j$ que cumpla la siguiente ecuación: $k\cdot j = 1\ (mod\ z)$, es decir, que $j$ sea el inverso multiplicativo de $k\ mod(n)$.  Luego la clave privada es el conjunto de los números $(n,j)$.</br>
Para cifrar un mensaje $p$ se debe usar la siguiente ecuación matemática: $p^k = e\ (mod\ n)$, donde e es el mensaje ya cifrado.</br>
Para descifrar el mensaje cifrado $e$ se debe usar la siguiente ecuación matemática: $e^j = p\ (mod\ n)$.</br>


# Implementación del algoritmo RSA en python:</br>
A continuación se mostrará una implementación del algoritmo RSA en python. Para efectos de practicidad, los números primos escogidos $p$ y $q$ no serán tan grandes como lo sugiere el algoritmo.</br>
Primero se procederá a la generación de las claves:
* Para realizar una prueba del funcionamiento, se cifrará el mensaje "COVID", y luego se descifrará.

```
import random
import math

#Función que revisa si un número es primo
def isPrime(number:int):
  if number < 2:
    return False

  for i in range(2,number):
    if number % i == 0:
      return False
  
  return True

#Esta función se basa en el postulado de Bertrand y regresa un número primo
#aleatorio entre n y 2n ingresados como parametro
def getPrime(number:int):
  if number % 2 == 0:
    number += 1

  result = random.randint(number,2*number)

  while not isPrime(result):
    result += 1
    if result == 2*number:
      result = number

  return result

#Algoritmo de euclides usando recursividad para encontrar el máximo común
#divisor
def gcdExtended(a, b):  
    
    if a == 0 :   
        return b, 0, 1
             
    gcd, x1, y1 = gcdExtended(b%a, a)  
     
    x = y1 - (b//a) * x1  
    y = x1  
     
    return gcd, x, y

#función que recibe el coeficiente de la clave publica y calcula el coeficiente
#de la clave privada
def get_d(e, phiEuler):
  divisor,x,y = gcdExtended(e, phiEuler)

  if x < 0:
    x = x % phiEuler

  return x

#función que halla todos los coprimos de phiEuler y selecciona uno de ellos como
#coeficiente de la clave publica
def  get_e(phiEuler):
  value = random.randint(3,phiEuler)
  while value < phiEuler:
    
    divisor,x,y = gcdExtended(value,phiEuler)
    if divisor == 1:
      return value
    value += 1
  return get_e(phiEuler)

#función que permite generar las llaves publicas y privadas
def generatePasswords():
  #Se generan 2 primos aleatorios distintos
  q = getPrime(1000)
  p = getPrime(3000)

  #Se calcula n como el producto entre p y q
  n = p*q

  #Se calcula φ(n) como (p-1)(q-1)
  phiEuler = (p-1)*(q-1)

  #Se define e, el cual es menor que φ(n) y es coprimo a φ(n)
  e = get_e(phiEuler)
  
  #Se define d como el inverso multiplicativo de e
  d = get_d(e,phiEuler)

  #Ahora, se define la clave pública como (n,e)
  publicKey = (n,e)

  #Ahora, se define la clave privada como (n,d)
  privateKey = (n,d)

  return publicKey,privateKey

def cipher(word,e,n):
  cipheredWord = [(ord(char) ** e) % n for char in word]
  return cipheredWord

def decipher(word,d,n):
  
  plain = [chr((char ** d) % n) for char in word]
  return ''.join(plain)

def binaryExp(base:int, exp:int):
  if exp == 1:
    return base
  elif exp % 2 == 0:
    return (base**(exp/2))**2
  elif exp % 2 != 0:
    return base*(base**(exp-1))

def RSA(wordToCipher):
  print('Bienvenido al sistema, calculando las claves para cifrado...')

  publicKey,privateKey = generatePasswords()

  print('las claves fueron generadas exitosamente y estan listas para usar.')
  print(f"{publicKey} {privateKey}")
  
  print("mensaje cifrado: ")
  cipheredMessage = cipher(wordToCipher,publicKey[1],publicKey[0])
  print(cipheredMessage)

  print("mensaje descifrado: ")
  decipheredMessage = decipher(cipheredMessage,privateKey[1],privateKey[0])
  print(decipheredMessage)

RSA("COVID")

```
